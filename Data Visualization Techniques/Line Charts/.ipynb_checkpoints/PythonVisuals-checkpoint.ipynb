```python
#Loading libaries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Reading in data
pop_df = pd.read_csv('world-population.csv')
```

### Line Chart

```python
plot1 = sns.lineplot(x='Year',y='Population', data = pop_df)
#Changing labels
plot1.set_title('World Population 1960 - 2009')
plot1.set_ylabel('Population (in billions)', fontsize=16)

#Style
sns.set_style('white')

#Removing Border
sns.despine()

#changing ticks
plt.ylim(3000000000,7000000000)
plt.locator_params(axis = 'y', nbins=5)

#Creating a text box for minor annotations
props = dict(boxstyle='round', facecolor='wheat', alpha=0.2)
plot1.text(1, 0.95,"*World Population has more than doubled since 1960 from 3 billion to 6.7 billion" , transform=plot1.transAxes, fontsize=14,
        verticalalignment='top', bbox=props)

plt.show()

```

### Step Chart

```python
#Creating step chart
plot2=sns.lineplot(x='Year',y='Population',data=pop_df,drawstyle='steps-pre')

#Style
sns.set_style('white')

plot2.set_title('World Population 1960 - 2009')
plot2.set_ylabel('Population (in billions)', fontsize=16)


#Removing Border
sns.despine()

#changing ticks
plt.ylim(3000000000,7000000000)
plt.locator_params(axis = 'y', nbins=5)

#Creating a text box for minor annotations
props = dict(boxstyle='round', facecolor='wheat', alpha=0.2)
plot2.text(1, 0.95,"*World Population appears to have even growth between 1960 and 2009" , transform=plot2.transAxes, fontsize=14,
        verticalalignment='top', bbox=props)

plt.show()
```
