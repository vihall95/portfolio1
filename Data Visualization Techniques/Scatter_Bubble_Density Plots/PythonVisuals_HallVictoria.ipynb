```python
#Importing packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### 1. Scatter Plot

```python
#loading data
crime_df = pd.read_csv("crimerates-by-state-2005.csv")
```

```python
#Checking dataframe
crime_df.head(5)
```

```python
#Checking for missing values
crime_df.isnull().sum()
```

```python
#creating scatterplot
sns.scatterplot(x='robbery',y='aggravated_assault', data=crime_df)
```

```python
print(crime_df[crime_df.aggravated_assault == crime_df.aggravated_assault.max()] )
```

```python
sns.scatterplot(x='robbery',y='aggravated_assault', data=crime_df)
plt.text(672.1,721.3,"Washington DC")
plt.title("Higher rates of robbery appear to correlate with an increase in assault", fontsize = 10)
plt.suptitle("2005 US States Crime Rates - Robbery vs. Aggravated Assault",y=1,fontsize=15)
plt.grid(False)
plt.show()
```

### 2. Bubble Chart

```python
#Using the same specifications of the scatter plot but including population as size of points
sns.scatterplot(x='robbery',y='aggravated_assault',size='population',sizes=(20,2000),legend=False, data=crime_df)
plt.text(672.1,721.3,"Washington DC")
plt.title("Bubble Size = State Population Size", fontsize = 10)
plt.suptitle("2005 US States Crime Rates - Robbery vs. Aggravated Assault",y=1,fontsize=15)
plt.grid(False)
plt.show()
```

### 3. Density Plot

```python
#Using kernal density plt
sns.kdeplot(data=crime_df,x='robbery')
plt.suptitle("2005 Robbery Rate Distributions",fontsize=15)
plt.text(750,.0003,"Washington DC")
plt.text(200,.005,"Most States rates are between 0-200")
```
