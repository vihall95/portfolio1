### Modeling

#### Random Forest

```python
X = train_df.drop(["price_range"],axis = 1)
y = train_df["price_range"]
```

```python
dummy_df =pd.get_dummies(X)
print(len(dummy_df.columns))
print(dummy_df.columns)
```

```python
from sklearn.model_selection import train_test_split
#Saving out a small set for validation - using kfold validation 
X_train,X_test, y_train, y_test = train_test_split(X,y,test_size=0.2, random_state=8675309)
```

```python
# evaluate random forest algorithm for classification
from numpy import mean
from numpy import std
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.ensemble import RandomForestClassifier
```

```python
#Looking at parameters
from pprint import pprint
model = RandomForestClassifier()

pprint(model.get_params())
```

```python
#Grid search to find hyperparameters
from sklearn.model_selection import RandomizedSearchCV
# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 200, stop = 2000, num = 10)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(10, 110, num = 11)]
max_depth.append(None)
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 4]
# Method of selecting samples for training each tree
bootstrap = [True, False]
# Create the random grid
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap}
pprint(random_grid)
```

```python
model = RandomForestClassifier()
# evaluate the model
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
n_scores = cross_val_score(model, X, y, scoring='accuracy', cv=cv, n_jobs=-1, error_score='raise',verbose = 10)
# report performance
print('Accuracy: %.3f (%.3f)' % (mean(n_scores), std(n_scores)))
```

```python
# Use the random grid to search for best hyperparameters
# First create the base model to tune
model = RandomForestClassifier()
# Random search of parameters, using 3 fold cross validation, 
# search across 100 different combinations, and use all available cores
model_random = RandomizedSearchCV(estimator = model, param_distributions = random_grid, n_iter = 100, cv = 3, verbose=10, random_state=42, n_jobs = -1)
# Fit the random search model
model_random.fit(X_train, y_train)
```

```python
model_random.best_params_
```

```python
#Evaluating Best random model
model = RandomForestClassifier(n_estimators = 400,min_samples_split =5,min_samples_leaf= 1,max_features= 'auto',
                              max_depth = 90, bootstrap = False)
# evaluate the model
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
n_scores = cross_val_score(model, X, y, scoring='accuracy', cv=cv, n_jobs=-1, error_score='raise',verbose = 10)
# report performance
print('Accuracy: %.3f (%.3f)' % (mean(n_scores), std(n_scores)))
```

```python
#Using results of random to get a good range for gridsearch
from sklearn.model_selection import GridSearchCV
# Create the parameter grid based on the results of random search 
param_grid = {
    'bootstrap': [True,False],
    'max_depth': [70,80, 90, 100, 110],
    'max_features': [2,3,4,5,6],
    'min_samples_leaf': [1,2,3,4,5],
    'min_samples_split': [3,5,7,9,11],
    'n_estimators': [100, 200, 300, 400, 1000]
}

# Create a based model
rf = RandomForestClassifier()
# Instantiate the grid search model
grid_search = GridSearchCV(estimator = rf, param_grid = param_grid, 
                           cv = 3, n_jobs = -1, verbose = 10)


grid_search.fit(X_train, y_train)
pprint(grid_search.best_params_)

best = grid_search.best_estimator_

# evaluate the model
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
n_scores = cross_val_score(best, X, y, scoring='accuracy', cv=cv, n_jobs=-1, error_score='raise',verbose = 10)
# report performance
print('Accuracy: %.3f (%.3f)' % (mean(n_scores), std(n_scores)))
```

```python
from sklearn.metrics import classification_report
#predicting on the test set
pred = best.predict(X_test)
print(classification_report(y_test,pred))
```

```python
feature_names = ['battery_power', 'blue', 'clock_speed', 'dual_sim', 'fc', 'four_g',
       'int_memory', 'm_dep', 'mobile_wt', 'n_cores', 'pc', 'px_height',
       'px_width', 'ram', 'sc_h', 'sc_w', 'talk_time', 'three_g',
       'touch_screen', 'wifi']

plt.barh(feature_names, best.feature_importances_)
plt.title("Random Forest Feature Importance")
plt.show()
```

```python
target_names = ['low','medium','high','very high']
```

```python
cr  = classification_report(y_test,pred,target_names = target_names,output_dict= True)
sns.color_palette("crest", as_cmap=True)
sns.heatmap(pd.DataFrame(cr).iloc[:-1,:].T, annot = True)
plt.title("Random Forest Classification Report")
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

#prepping data for model training and testing
X = train_df.drop(["price_range"],axis = 1)
y = train_df["price_range"]
X_train,X_test, y_train, y_test = train_test_split(X,y,test_size=0.2, random_state=8675309)


cat_attribs = ['blue', 'dual_sim', 'four_g', 'three_g', 'touch_screen', 'wifi']
full_pipeline = ColumnTransformer([('cat', OneHotEncoder(handle_unknown='ignore'), cat_attribs)], remainder='passthrough')
encoder = full_pipeline.fit(X_train)
X_train = encoder.transform(X_train)
X_test = encoder.transform(X_test)
```

```python
from xgboost import XGBClassifier
#initiating model. 
estimator = XGBClassifier(
    objective= 'multi:softmax',
    seed=42
)
```

```python
#Base Model
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score, KFold



estimator.fit(X_train,y_train)
kfold = KFold(n_splits=3, shuffle=True)
kf_cv_scores = cross_val_score(estimator, X_train, y_train, cv=kfold,verbose = 10 )
print("K-fold CV average score: %.2f" % kf_cv_scores.mean())
```

```python
#setting parameters for gridsearch
parameters = {
    'max_depth': range (2,5,1),
    'n_estimators': range(80, 200, 40),
    'learning_rate': [0.1, 0.01, 0.05],
    'min_child_weight':[1,5,10],
    'gamma': [0.5,1,1.5,2,5],
    'subsample': [0.6, 0.8, 1.0],
    'colsample_bytree': [0.6, 0.8, 1.0]
    
}
```

```python
# Instantiate the grid search model
from xgboost import XGBClassifier
#initiating model. 
estimator = XGBClassifier(
    objective= 'multi:softmax',
    seed=42
)
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

#prepping data for model training and testing
X = train_df.drop(["price_range"],axis = 1)
y = train_df["price_range"]
X_train,X_test, y_train, y_test = train_test_split(X,y,test_size=0.2, random_state=8675309)


cat_attribs = ['blue', 'dual_sim', 'four_g', 'three_g', 'touch_screen', 'wifi']
full_pipeline = ColumnTransformer([('cat', OneHotEncoder(handle_unknown='ignore'), cat_attribs)], remainder='passthrough')
encoder = full_pipeline.fit(X_train)
X_train = encoder.transform(X_train)
X_test = encoder.transform(X_test)



grid_search = GridSearchCV(estimator = estimator, param_grid = parameters, 
                           cv = 3, n_jobs = -1, verbose = 10)


grid_search.fit(X_train, y_train)
pprint(grid_search.best_params_)

best = grid_search.best_estimator_

# evaluate the model
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
n_scores = cross_val_score(best, X_train, y_train, scoring='accuracy', cv=cv, n_jobs=-1, error_score='raise',verbose=10)
# report performance
print('Accuracy: %.3f (%.3f)' % (mean(n_scores), std(n_scores)))
```

```python
from xgboost import plot_importance
orig_feature_names = ['battery_power', 'clock_speed', 'fc', 'int_memory', 'm_dep',
       'mobile_wt', 'n_cores', 'pc', 'px_height', 'px_width', 'ram', 'sc_h',
       'sc_w', 'talk_time', 'blue_0', 'blue_1', 'dual_sim_0', 'dual_sim_1',
       'four_g_0', 'four_g_1', 'three_g_0', 'three_g_1', 'touch_screen_0',
       'touch_screen_1', 'wifi_0', 'wifi_1']


best.get_booster().feature_names = orig_feature_names
plot_importance(best, max_num_features = 15)
plt.show()
```

```python
from sklearn.metrics import classification_report
#predicting on the test set
pred = best.predict(X_test)
print(classification_report(y_test,pred))
```

```python
cr  = classification_report(y_test,pred,target_names = target_names,output_dict= True)

sns.heatmap(pd.DataFrame(cr).iloc[:-1,:].T, annot = True)
plt.title("XGBoost Classification Report")
```
